```python
#импорт необходимых библиотек
import sys
sys.path.append(r'c:\users\alexey\desktop\prof_intro\py_ml\venv\lib\site-packages')

from PIL import Image
import numpy as np
```

```python
#открытие изображения при помощи метода open()
img=Image.open(r"C:\Users\Alexey\Desktop\ВУЗ_ДОКИ\photo.jpg").resize((300, 300))
img
```

```python
#создание массива для разложения изображения по каналам
img_arr=np.array(img, dtype=np.float32)

red_col = img_arr[:,:,0]
green_col =img_arr[:,:,1]
blue_col = img_arr[:,:,2]
```

```python
#SVD разложение при помщи функции np.linalg.svd()

U_r, S_r, V_r = np.linalg.svd(red_col, full_matrices = False)
U_b, S_b, V_b = np.linalg.svd(blue_col, full_matrices = False)
U_g, S_g, V_g = np.linalg.svd(green_col, full_matrices = False)
```

```python
#замена значений для приведения к типу uint8
def prep_to_convert(arr):
    arr[arr<0], arr[arr>255]= 0, 255
    return arr
    

for r in range (1, 150, 30):
    Y_red = U_r[:,:r] @ (np.diag(S_r[:r])) @ (V_r[:r,:])
    Y_red= prep_to_convert(Y_red)
    Y_red = Y_red.reshape(300, 300, 1)
    
    
    Y_green = U_g[:,:r]  @ (np.diag(S_g[:r])) @ (V_g[:r,:]) 
    Y_green=prep_to_convert(Y_green)
    Y_green = Y_green.reshape(300, 300, 1)
    
    Y_blue = U_b[:,:r] @ (np.diag(S_b[:r])) @ (V_b[:r,:])
    Y_blue=prep_to_convert(Y_blue)
    Y_blue = Y_blue.reshape(300, 300, 1)
    
    #Функция concatenate() соединяет массивы вдоль указанной оси
    Y = np.concatenate((np.concatenate((Y_red, Y_green), axis = 2), Y_blue), axis = 2)
    #получение изображения при помощи метода fromarray() и вывод при помощи метода show()
    Image.fromarray(np.asarray(Y, dtype=np.uint8)).show()
```
